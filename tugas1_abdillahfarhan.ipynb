### TAHAP 5
### ENCODING LABEL

```python
from sklearn.preprocessing import LabelEncoder
import numpy as np
```

```python
cols = ["Nama", "Jenis_Kelamin", "Prodi", "Status", "Nilai_Akhir"]

for col in cols:
    df[col] = df[col].astype(str).str.strip()
```

```python
encoders = {}

for col in cols:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])
    encoders[col] = le
```

```python
df.head(20)
```

```python
df = df.drop(columns=["Nama", "ID"])
```

```python
df["Tahun"] = df["Tanggal_Ujian"].dt.year
df["Bulan"] = df["Tanggal_Ujian"].dt.month
df["Hari"] = df["Tanggal_Ujian"].dt.day
```

```python
df = df.drop(columns=["Tanggal_Ujian"])
```

### TAHAP 7
### VISUALISASI DATA

```python
import matplotlib.pyplot as plt
```

```python
mean_age = df['Umur'].mean()

plt.figure(figsize=(8,5))

counts, bins, patches = plt.hist(
    df['Umur'],
    bins=10,
    color='#4ECDC4',
    edgecolor='black'
)

for count, patch in zip(counts, patches):
    plt.text(
        patch.get_x() + patch.get_width()/2,
        count,
        int(count),
        ha='center',
        va='bottom'
    )

plt.axvline(mean_age, linestyle='--')
plt.text(mean_age, max(counts)*0.9, f'Rata-rata: {mean_age:.1f}', rotation=0)

plt.title('Distribusi Umur Mahasiswa')
plt.xlabel('Umur')
plt.ylabel('Frekuensi')

plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
prodi_labels = encoders['Prodi'].inverse_transform(df['Prodi'])
prodi_counts = pd.Series(prodi_labels).value_counts()


plt.figure(figsize=(8,5))

bars = plt.bar(
    prodi_counts.index,
    prodi_counts.values,
    color='#4ECDC4',
    edgecolor='black',
    alpha=0.8,
    width=0.6
)

for bar in bars:
    plt.text(
        bar.get_x() + bar.get_width()/2,
        bar.get_height(),
        int(bar.get_height()),
        ha='center',
        va='bottom'
    )

plt.title('Jumlah Mahasiswa per Prodi', fontsize=10)
plt.xlabel('Prodi', fontsize=12)
plt.ylabel('Jumlah Mahasiswa', fontsize=10)

plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()

```
